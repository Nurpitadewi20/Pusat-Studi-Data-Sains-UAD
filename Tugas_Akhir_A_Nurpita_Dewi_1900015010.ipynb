## Data Preparation

```
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```
#Membaca file
ramen_df = pd.read_csv('/ramen-ratings.csv')
#Keterangan isi index
print(ramen_df.keys())
ramen_df
```

1. Bahan atau rasa apa yang sering diiklankan pada label paket ramen.?

```
men_var_df = pd.DataFrame(ramen_df[['Variety', 'Stars']])
men_var_df.sort_values(by=['Stars'])
```

2. Bagaimana peringkat ramen dibandingkan dengan peringkat untuk produk makanan lain.

```
men_brand_df = pd.DataFrame(ramen_df[['Brand','Stars']])
men_brand_df.sort_values(by=['Stars'])
```

3. Bagaimana produksi ramen didistribusikan secara internasional?

```
men_count_df = pd.DataFrame(ramen_df[['Brand','Country']])
men_count_df.sort_values(by=['Country'])
```

## Data Visualization

```
gbr_iklan = px.bar(men_var_df,
               x = 'Variety',
               y = 'Stars')

gbr_iklan.update_layout(
    title = 'Diagram batang variasi bahan yang diiklankan ',
    title_font_color = 'blue',
    title_font_size = 30,

    xaxis = dict(
        title = 'Variety',
        titlefont_size = 16,
        tickfont_size = 9),
    
    yaxis = dict(
        title = 'Stars',
        titlefont_size = 16,
        tickfont_size = 14))

gbr_iklan.show()
```

```
gbr_peringkat = px.bar(men_brand_df,
               x = 'Brand',
               y = 'Stars')

gbr_peringkat.update_layout(
    title = 'Diagram batang Peringkat Produk ',
    title_font_color = 'blue',
    title_font_size = 30,

    xaxis = dict(
        title = 'Brand',
        titlefont_size = 16,
        tickfont_size = 9),
    
    yaxis = dict(
        title = 'Stars',
        titlefont_size = 16,
        tickfont_size = 14))

gbr_peringkat.show()
```
